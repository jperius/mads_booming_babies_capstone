<h2>Recall Optimized Modeling Visualization</h2>

This notebook was used to generate visualizations for our report.

```python
import pandas as pd
import plotly.graph_objects as go
import numpy as np
```

```python
manual_df = pd.DataFrame({
    'model': ['Dummy - Uniform', 'Dummy - Stratified','XGBoost Classifier DB: 0.5', 'Logistic Regression DB: 0.5', 'Logistic Regression DB: 0.3'],
    'recall': [0.5002336022164544, 0.49956221518056315, 0.5834563089495611, 0.6234084076665803, 1.0],
    'color': ['#A8A9AD', '#7C7D81','#6c94b4', '#1c416d', '#d9773d']
})

manual_df
```

```python
def wrap_labels(labels, width=10):
    wrapped = []
    for label in labels:
        words = label.split(" ")
        lines = []
        current = []

        for word in words:
            # test if adding the next word exceeds width
            test = " ".join(current + [word])
            if len(test) > width:
                lines.append(" ".join(current))
                current = [word]
            else:
                current.append(word)

        if current:
            lines.append(" ".join(current))

        wrapped.append("<br>".join(lines))

    return wrapped


fig = go.Figure(
    data=[
        go.Bar(
            x=manual_df["model"],
            y=manual_df["recall"],
            marker=dict(color=manual_df["color"]),
            text=np.round(manual_df["recall"], 2),
            textposition="outside",
            textfont=dict(size=14)
        )
    ],
    layout=go.Layout(
        width=1000,
        height=500,
        #title="Recall Comparison",
        xaxis_title="Model",
        yaxis_title="Recall",
        plot_bgcolor='rgba(0,0,0,0)',
        paper_bgcolor='rgba(0,0,0,0)',
        yaxis=dict(
            range=[0, 1.1],
            showgrid=False,
            showticklabels=False,
            title='',
            zeroline=False
        ),
        xaxis=dict(
        tickmode="array",
        #ticktext=wrap_labels(manual_df["model"], 10),
        tickvals=manual_df["model"]
        )
    )
)

fig.show()
```
