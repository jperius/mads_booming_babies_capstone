```python
with open(BASE_DIR / 'variable_selection' / 'var_name_map_optimizer.json', 'r') as f:
    var_name_map = json.load(f)
var_name_map
```

```python
inverse_var_name_map = {}

for key, values in var_name_map.items():
    for v in values:
        inverse_var_name_map.setdefault(v, []).append(key)
```

```python
all_features_condensed = sorted(list(set(
    k for f in all_features for k in inverse_var_name_map[f]
)))

trial1_features_condensed = sorted(list(set(
    k for f in trial1_features for k in inverse_var_name_map[f]
)))

trial2_features_condensed = sorted(list(set(
    k for f in trial2_features for k in inverse_var_name_map[f]
)))

trial3_features_condensed = sorted(list(set(
    k for f in trial3_features for k in inverse_var_name_map[f]
)))

trial4_features_condensed = sorted(list(set(
    k for f in trial4_features for k in inverse_var_name_map[f]
)))

print(len(all_features_condensed))
print(len(trial1_features_condensed) / len(all_features_condensed))
print(len(trial2_features_condensed) / len(all_features_condensed))
print(len(trial3_features_condensed) / len(all_features_condensed))
print(len(trial4_features_condensed) / len(all_features_condensed))
```

```python
model_features = {
    'Trial 1': trial1_features_condensed,
    'Trial 2': trial2_features_condensed,
    'Trial 3': trial3_features_condensed,
    'Trial 4': trial4_features_condensed,
}

rows = []
for model, feats in model_features.items():
    rows.append(
        { 'Model': model, **{f: (f in feats) for f in all_features_condensed} }
    )

df_viz = pd.DataFrame(rows)

df_viz
```

```python
df_viz = df_viz.sort_values("Model", ascending=False).reset_index(drop=True)
mask = df_viz.set_index("Model")

fig = go.Figure(
    data=go.Heatmap(
        z=mask.values.astype(int),
        x=mask.columns,
        y=mask.index,
        colorscale=[[0, "rgba(0,0,0,0)"], [1, "#1c416d"]],
        showscale=False,
        xgap=3,
        ygap=3 
    )
)

fig.update_layout(
    width=1200,
    height=400,
    xaxis_title="Variable",
    yaxis_title="Model",
    plot_bgcolor='rgba(0,0,0,0)',
    paper_bgcolor='rgba(0,0,0,0)',
    yaxis=dict(
        showgrid=False,
        #showticklabels=False,
        zeroline=False
    ),
    xaxis=dict(
        showgrid=False,
        #showticklabels=False,
        zeroline=False,
        tickangle=45
    ),
)

fig.show()
```
