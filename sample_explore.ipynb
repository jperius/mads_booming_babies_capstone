```python
import pandas as pd
import numpy as np
from pathlib import Path

BASE_DIR = Path().resolve()
```

```python
df_2024 = pd.read_csv(BASE_DIR / "data" / "natality2024us.csv", nrows=1)
df_2024_cols = df_2024.columns.to_list()
```

```python
for file in BASE_DIR.glob("data/*.csv"):

    print(f'Loading data from {file}')
    df = pd.read_csv(file)
    sampled_df = df.sample(n=10000, random_state=42)
    sampled_df = sampled_df.reindex(columns=df_2024_cols)

    out_file_path = BASE_DIR / "sampled_natality_data.csv"

    file_exists = out_file_path.exists()

    with open(out_file_path, 'a', encoding='utf-8') as f:
        sampled_df.to_csv(f, header=not file_exists, index=False)
```
